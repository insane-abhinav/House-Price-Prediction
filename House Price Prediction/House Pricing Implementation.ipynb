```python
oceanProximity = pd.get_dummies(df['ocean_proximity'],drop_first=True)
```

```python
oceanProximity
```

```python
df = df.drop('ocean_proximity',axis=1)
```

```python
df
```

```python
df = pd.concat([df,oceanProximity],axis=1)
```

```python
df
```

## Handling Missing Value

```python
bmean = df['total_bedrooms'].mean()
```

```python
df['total_bedrooms'].fillna(value=bmean,inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

## Dependent and Independent feature Splitting

```python
y=df['median_house_value'].values
```

```python
X = df.drop(labels='median_house_value',axis=1).values
```

```python
X
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(X_train,y_train)
```

```python
pred = reg.predict(X_test)
```

## R2 Score

```python
reg.score(X_test,y_test)
```

## RMSE

```python
from sklearn.metrics import mean_squared_error
import math
mse=mean_squared_error(y_test,pred)
rmse=math.sqrt(mse)
rmse
```

## Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,pred)
```
